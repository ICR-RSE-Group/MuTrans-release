# MuTrans Analysis of Epithelial-Mesenchymal Transition (EMT)

## Load the dataset and explore with Scanpy

```python
import pandas as pd
import scanpy as sc
sc.settings.set_figure_params(dpi=100, frameon=False, figsize=(3, 3), facecolor='white')
import matplotlib.pyplot as plt
import pyMuTrans as pm

datadir = "../Data/"
adata = sc.read_csv(datadir+'data_emt.csv')
genes = pd.read_csv(datadir+'emt_genes.csv')
adata.var_names = genes['genes']
```

```python
sc.pp.log1p(adata)
sc.pp.scale(adata)
sc.tl.tsne(adata,perplexity=40)
```

Using leiden algorithm to detect clusters (with different resolutions)

```python
sc.pp.neighbors(adata)
sc.tl.leiden(adata,resolution = 0.3)
sc.pl.tsne(adata, color=['leiden','Epcam','Mmp19'])
```

```python
sc.tl.leiden(adata,resolution = 0.5)
sc.pl.tsne(adata, color=['leiden','Epcam','Mmp19'])
```

## Estimate Cluster Number with EPI (Eigen-Peak Index)

```python
fig = plt.figure(figsize=(6, 3))
par = {"choice_distance":"cosine"}
out = pm.plot_cluster_num(adata, par, k_plot= 10) #look for the peaks
```

## MuTrans Analysis Based on the AnnData Object

```python
par = {"choice_distance":"cosine","perplex":150.0,"K_cluster":3.0,"trials":10,"reduction_coord":'tsne',"write_anndata":True} 
adata = pm.dynamical_analysis(adata,par)
```

```python
sc.pl.tsne(adata, color=['attractor','entropy','Epcam','Mmp19'])
```

```python
sc.tl.rank_genes_groups(adata, 'attractor', method='wilcoxon')
sc.pl.rank_genes_groups(adata, n_genes=10, sharey=False)
```

```python
sc.pl.tsne(adata, color=['Col1a1','ERCC.21','Krt5'])
```

Infer the transition trajectories from Epithelial to Mesenchymal, with the MPPT (most probable path tree) approach

```python
fig = plt.figure(figsize=(8, 8))
pm.infer_lineage(adata,si=2,sf=0,method = "MPPT",size_point =40,alpha_point = 0.5)
```

In the EPI figure, we actually observe another peak at K=5. What if we choose the parameter like this?

```python
par['K_cluster'] = 5.0
adata = pm.dynamical_analysis(adata,par)
```

```python
sc.pl.tsne(adata, color=['attractor','entropy','Epcam','Mmp19'])
```

```python
sc.tl.rank_genes_groups(adata, 'attractor', method='wilcoxon')
sc.pl.rank_genes_groups(adata, n_genes=10, sharey=False)
```

```python
fig = plt.figure(figsize=(8, 8))
pm.infer_lineage(adata,si=3,sf=4,method = "MPPT",size_point =30, alpha_point = 0.5)
```

Note that two new attractors appear (both are ICS states). However, we observe that the qualitative conclusion about transition trajectory remains relatively robust, indicating that roles of ICS.
