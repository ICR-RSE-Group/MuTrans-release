# Analysis of the Myelopoiesis Dataset from Olsson et al.

```python
import pandas as pd
import numpy as np
import scanpy as sc
sc.settings.set_figure_params(dpi=100, frameon=False, figsize=(3, 3), facecolor='white')
import matplotlib.pyplot as plt
import pyMuTrans as pm

datadir = "../Data/"
adata = sc.read_csv(datadir+'data_olsson.csv') # load the pre-processed data
genes = pd.read_csv(datadir+'genes_olsson.csv',header=None)
label = pd.read_csv(datadir+'label_olsson.csv',header= None)
adata.var_names = genes.iloc[:,0]
adata.obs['olsson_labels'] = label.values
```

```python
sc.pp.neighbors(adata,metric = 'cosine',n_neighbors=10, use_rep='X')
sc.tl.umap(adata)
adata.obs['olsson_labels'] =  adata.obs['olsson_labels'].astype('str')
sc.pl.umap(adata, color = ['olsson_labels'])
```

## Plotting EPI

```python
fig = plt.figure(figsize=(6, 3))
par = {"choice_distance":"euclid"}
out = pm.plot_cluster_num(adata, par, k_plot= 15) #look for the peaks
```

## MuTrans Analysis

```python
par = {"choice_distance":"cosine","perplex":150.0,"K_cluster":10.0,"trials":100,"reduction_coord":'umap',"weight_scale":False,"write_anndata":True} 
adata = pm.dynamical_analysis(adata,par)
```

```python
import seaborn as sns
color_palette = sns.color_palette('Set3', 10)
sc.pl.umap(adata, color=['attractor','entropy'],vmax = 'p95',palette = color_palette)
```

## Transition Path Analysis

Global lineage inferred by MPFT (maximum probability flows)

```python
fig = plt.figure(figsize=(10, 6))
pm.infer_lineage(adata,method = "MPFT",flux_fraction = 0.3,size_point =40,alpha_point = 0.5,size_text = 15,color_palette=color_palette)
```

Transition path from HSPC to Eryth

```python
fig = plt.figure(figsize=(10, 6))
pm.infer_lineage(adata,si=0,sf=4,method = "MPPT",flux_fraction = 0.3,size_point =40,alpha_point = 0.5,size_text = 15,color_palette=color_palette)
```

Transition path from HSPC to Meg

```python
fig = plt.figure(figsize=(10, 6))
pm.infer_lineage(adata,si=0,sf=3,method = "MPPT",flux_fraction = 0.3,size_point =40,alpha_point = 0.5,size_text = 15,color_palette=color_palette)
```
